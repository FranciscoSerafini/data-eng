### Realizamos ETL para nuestros datos

```python
#hacemos una copia de los dataframe
df_amazon = amazon.copy()
df_disney = disney.copy()
df_netflix = netflix.copy()
df_hulu = hulu.copy()
```

```python
df_amazon.head()
```

```python
#agreamos columnas para asginar el valor de la plataforma a la que pertenencen
df_amazon['Plataforma']='Amazon'
df_disney['Plataforma']='Disney'
df_netflix['Plataforma']='Netflix'
df_hulu['Plataforma'] = 'Hulu'
```

```python
df_amazon.head()
```

```python
df_netflix.head()
```

```python
#remplazamos los campos vacios en las columna duration 
df_netflix['duration'].fillna('sin datos', inplace=True)
df_hulu['duration'].fillna('sin datos', inplace=True)
```

```python
df_netflix
```

```python
df_hulu
```

```python
#debemos separar las columnas duration en dos
def separar_columnas(df):
   #creamos listas vacias para alojar los resultados
    cantidad_temporadas = []
    duracion = []
    
    #recorremos las columnas
    for i in df['duration']:
        if isinstance(i,str)and'Season' in i:
            cantidad_temporadas.append(i.split()[0])
        else:
            cantidad_temporadas.append('0')
    
        if isinstance(i,str) and 'min' in i:
            duracion.append(i.split()[0])
        else:
            duracion.append('0')
            
    df['cantidad_temporadas'] = cantidad_temporadas
    df['duracion'] = duracion
    return df
```

```python
#asignamos a cada df
df_amazon = separar_columnas(df_amazon)
df_disney = separar_columnas(df_disney)
df_hulu = separar_columnas(df_hulu)
df_netflix = separar_columnas(df_netflix)
```

```python
df_netflix.head()
```

```python
#eliminamos la columna duracion ya que la separamos en 2
df_amazon.drop(columns='duration', inplace=True)
df_disney.drop(columns='duration', inplace=True)
df_hulu.drop(columns='duration', inplace=True)
df_netflix.drop(columns='duration', inplace=True)
```

```python
df_netflix.head()
```

```python
#concateneacion de los dataframe
df_plataformas = pd.concat([df_amazon,df_disney,df_hulu,df_netflix], axis=0)
```
